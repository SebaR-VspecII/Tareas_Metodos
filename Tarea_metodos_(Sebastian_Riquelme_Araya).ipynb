```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
    return x * np.sin(x)

def taylor_f(x, n_terms = 10):
  taylor_sum = 0
  for n in range(n_terms):
    term = ((-1)**n * x**(2*n + 2)) / math.factorial(2*n + 1)
    taylor_sum += term
  return taylor_sum

x = np.linspace(-10, 10, 500)

y_real = f(x)
y_taylor = taylor_f(x)


plt.plot(x, y_real, label=r'$f\ (x) = xsin(x)$', color = 'green')
plt.plot(x, y_taylor, label=r'$Serie\ de\ Taylor$', linestyle='--', color = 'orange')
plt.xlabel(r'$x$')
plt.ylabel(r'$f\ (x)\ &\ Su\ Serie\ de\ Taylor$')
plt.title(r'$Grfica\ de\ xsin(x)\ &\ Su\ Serie\ de\ Taylor$')
plt.legend()
plt.grid(True)
plt.show()

# Hasta cul valor de x parece ser una buena aproximacin?
# Hasta el valor de x=[-7.5 , 7.5], ya que, luego la serie de Taylor difiere completamente de la funcin, en cambio dentro del intervalo sealado la serie de Taylor cumple con ser una buena aproximacin de  la funcin.
```
